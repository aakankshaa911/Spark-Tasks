### Task : Create a decision Tree Classifier and Visualize it graphically

##### Importing libraries

```python
# importing libraries
from IPython import display

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn import tree
```

#### Importing warnings

```python
import warnings
warnings.filterwarnings('ignore')
```

#### Working with the data

```python
df = pd.read_csv(r'C:\Users\Aakanksha\Documents\IRIS_Dataset.csv')
df
```

```python
df = df.iloc[: , 1:]
df
```

```python
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
print(X)
```

```python
y = load_iris().target
y
```

```python
iris = load_iris()
```

#### Splitting the data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

####  Fit the classifier

```python
clf = tree.DecisionTreeClassifier(max_depth=3, random_state=10)
clf.fit(X_train, y_train)
```

#### Plotting the decision tree classifier

```python
fig = plt.figure(figsize=(20,18))
tree.plot_tree(clf)
plt.show()
```

```python
fig = plt.figure(figsize=(20,18))
tree.plot_tree(clf,
               feature_names = iris.feature_names, 
               class_names=iris.target_names,
               rounded=True, 
               filled = True)
plt.show()
```

#### Checking the accuracy score on test set

```python
clf.score(X_test, y_test)
```
